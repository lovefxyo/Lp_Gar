```python
import pandas as pd
```

```python
#define settings
path="./graph_dataset/dblp_1980/"
graph_e_data="dblp_e.csv"
graph_v_data="dblp_v_new.csv"
graph_route="wenzhi_delta_vertex_dbpedia_g_5.txt"
result="/pykeen/top_100000_testing_new.csv"
e_data=pd.read_csv(path+graph_e_data,delimiter=",")
v_data=pd.read_csv(path+graph_v_data,delimiter=",")
predict=pd.read_csv(path+result,delimiter=",")
```

```python
#结果与label对应上
predict_new=pd.merge(pd.merge(predict,v_data,left_on="head_label",right_on="vertex_id:int"),v_data,left_on="tail_label",right_on="vertex_id:int")
predict_new.reindex(columns=["head_label","relation_label","tail_label","label_id:int_x","label_id:int_y","score","name:string_x","name:string_y"]).to_csv(path+"/pykeen/top_100000_final.csv")
predict_triple=predict_new.reindex(columns=["label_id:int_x","relation_label","label_id:int_y"])
predict_triple.value_counts().to_csv(path+"/pykeen/top_100000_group.csv")
```

```python
#原图与label对应上
graph_new=pd.merge(pd.merge(e_data,v_data,left_on="source_id:int",right_on="vertex_id:int"),v_data,left_on="target_id:int",right_on="vertex_id:int")
graph_new.reindex(columns=["edge_id:int","source_id:int","label_id:int","target_id:int","vertex_id:int_x","label_id:int_x","vertex_id:int_y","label_id:int_y"]).to_csv(path+"/3_yago_full.csv")
graph_new=pd.read_csv(path+"3_yago_full.csv")
graph_triple=graph_new.reindex(columns=["label_id:int_y","label_id:int_x","label_id:int"])
graph_triple
graph_new
```
