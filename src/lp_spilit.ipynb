```python
#method 1
lp_path="./graph_dataset/dblp_1980/pykeen/top_100000_final.csv"
lp_file=pd.read_csv(lp_path)
lp_file=lp_file.reindex(columns=["head_label","relation_label","tail_label","label_id:int_x","label_id:int_y"])
lp_group=lp_file.groupby(["label_id:int_x","relation_label","label_id:int_y"])
for name,group in lp_group:
    group.to_csv("./graph_dataset/dblp_1980/pykeen/top_100000_group/"+str(name).replace(' ','')+".csv",index=None)
```
